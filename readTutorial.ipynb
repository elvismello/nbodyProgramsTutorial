# Read Snapshots

This notebook has the pourpose of exibiting how to read snapshots using various codes.

```python
# general libraries
import numpy as np
from scipy import stats

# writting/reading libraries
import unsio.input as uns_in
import h5py

# plotting
import matplotlib
from matplotlib import pyplot as plt
```

```python
SNAPSHOT_NAME_G2 = "snapshots/snapshotG2"
SNAPSHOT_NAME_G3 = "snapshots/snapshotG3.hdf5"
```

## Gadget 2

### UNSIO

```python
snapshot_object = uns_in.CUNS_IN(SNAPSHOT_NAME_G2)
snapshot_object.nextFrame() # essential

_, positions = snapshot_object.getData("disk", "pos")


# positions are returned in a list where they are organized as xyzxyz...xyz
# reshaping positions
positions = np.array([positions[0::3], positions[1::3], positions[2::3]]).T
```

```python
positions
```

```python
# plotting

# using subplots enviroment
fig, axs = plt.subplots(2, 2, figsize=(12, 12))


# setting each graph

# changing projection from the fourth plot
axs[1, 1].remove()
axs[1, 1] = fig.add_subplot(2, 2, 4, projection="3d")


# setting data
axs[0, 0].scatter(positions[:, 0], positions[:, 1], s=0.1)
axs[1, 0].scatter(positions[:, 0], positions[:, 2], s=0.1)
axs[0, 1].scatter(positions[:, 1], positions[:, 2], s=0.1)
axs[1, 1].scatter(positions[:, 0], positions[:, 1], zs=positions[:, 2], s=0.1)


# setting limits
axs[0, 0].set_xlim((-40, 40))
axs[1, 0].set_xlim((-40, 40))
axs[0, 1].set_xlim((-40, 40))
axs[1, 1].set_xlim((-40, 40)) # 3d graph

axs[0, 0].set_ylim((-40, 40))
axs[1, 0].set_ylim((-40, 40))
axs[0, 1].set_ylim((-40, 40))
axs[1, 1].set_ylim((-40, 40)) # 3d graph

axs[1, 1].set_zlim((-40, 40)) # 3d graph
```

### H5PY

```python
# getting coordinates from the snapshot
with h5py.File(SNAPSHOT_NAME_G3) as snapshot_object:
    print(snapshot_object.keys())
    
    positions = snapshot_object["PartType2"]["Coordinates"][:]
```

```python
# plotting

# using subplots enviroment
fig, axs = plt.subplots(2, 2, figsize=(12, 12))


# setting each graph

# changing projection from the fourth plot
axs[1, 1].remove()
axs[1, 1] = fig.add_subplot(2, 2, 4, projection="3d")


# setting data
axs[0, 0].scatter(positions[:, 0], positions[:, 1], s=0.1)
axs[1, 0].scatter(positions[:, 0], positions[:, 2], s=0.1)
axs[0, 1].scatter(positions[:, 1], positions[:, 2], s=0.1)
axs[1, 1].scatter(positions[:, 0], positions[:, 1], zs=positions[:, 2], s=0.1)


# setting limits
axs[0, 0].set_xlim((-40, 40))
axs[1, 0].set_xlim((-40, 40))
axs[0, 1].set_xlim((-40, 40))
axs[1, 1].set_xlim((-40, 40)) # 3d graph

axs[0, 0].set_ylim((-40, 40))
axs[1, 0].set_ylim((-40, 40))
axs[0, 1].set_ylim((-40, 40))
axs[1, 1].set_ylim((-40, 40)) # 3d graph

axs[1, 1].set_zlim((-40, 40)) # 3d graph
```

## Plots

```python
with h5py.File(SNAPSHOT_NAME_G3) as snapshot_object:
    positions = snapshot_object["PartType0"]["Coordinates"][:]
    densities = snapshot_object["PartType0"]["Density"][:]


```

```python
statistic, x_edge, y_edge, binnumber = stats.binned_statistic_2d(positions[:, 0], positions[:, 1], densities, bins=196)

statistic[np.isnan(statistic)] = 0

plt.imshow(statistic, cmap="inferno")
```

```python
ls = matplotlib.colors.LightSource(azdeg=15, altdeg=45)
cmap = plt.cm.bone

plt.imshow(ls.hillshade(statistic, vert_exag=40), cmap=cmap)

plt.savefig("test.pdf")
```
